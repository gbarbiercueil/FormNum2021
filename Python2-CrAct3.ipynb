<center> Proposition de correction

```python
%matplotlib notebook
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import linregress

# Définition des listes
Vb=[]
Vb1=[]
Vb2=[]     
s=[] 
s1=[]    
s2=[]
na=[]
nb=[]
nc=[]

# Lecture d'un fichier csv pour en faire une liste 
                # le ficher csv est dans le même dossier que le programme
                # la première ligne du fichier contient les intitulés de chaque colonne ; ici "t (s);x (m);y (m)"
                # de la deuxième à la dernière ligne, on trouve les valeurs des grandeurs séparés par des ;
                # le contenu des lignes est considéré comme du texte
with open('titrage.csv','r') as fichier:        # le contenu du csv est placé dans fichier
    fichier.readline()              # on lit la première ligne de fichier (intitulés non utilisés)
    for ligne in fichier:               # a chaque passage, ligne prend la valeur de la nouvelle ligne de fichier
        ligne_lue=ligne.split(';')      # ligne_lue est une liste des valeurs de la variable ligne séparées par ';'
        Vb.append(float(ligne_lue[0].replace(',','.')))  # append signifie qu'on ajoute à la liste t
        s.append(float(ligne_lue[1].replace(',','.')))  # replace change la "," par un "."

N=len(Vb)

# On sépare les conductivités en deux listes, avant et après la rupture de pente
i=0
while s[i+1]<s[i]:
    Vb1.append(Vb[i])
    s1.append(s[i])
    na.append(0.00103-(0.1*Vb[i]/1000))     # Réactif en excès
    nb.append(0)                            # Réactif limitant
    nc.append(0.1*Vb[i]/1000)               # Calcul de la quantité du produit (l'autre est le solvant)
    i=i+1

i=i+1                       # On evite de garder la valeur autour de la rupture de pente
while i<N :
    Vb2.append(Vb[i])
    s2.append(s[i])
    na.append(0)                            # Réactif limitant
    nb.append((0.1*Vb[i]/1000)-0.00103)     # Réactif en excès
    nc.append(0.00103)                      # La quantité de produit est maximale
    i=i+1
 
 # on fait la régression des deux listes   
a1,b1,ro1,_,_=linregress(Vb1,s1)
a2,b2,ro2,_,_=linregress(Vb2,s2)
x_inter=(b2-b1)/(a1-a2)
y_inter=a1*x_inter+b1
print('Volume équivalent : ',x_inter,' mL')

# Calcul de la concentration Ca
C_b=0.1
V_eq=x_inter
V_a=20
C_a=C_b*V_eq/V_a
print('La concentration en acide sulfamique est : ',C_a,' mol/L')

# Calcul de la quantité de soluté na
n_a=C_b*V_eq/1000
print('La quantité de soluté est : ',n_a,' mol')

# Bornes du modèle
x_min=min(Vb1)
x_max=max(Vb2)

# Abscisses et ordonnées du modèle
x_modele=np.linspace(x_min,x_max,101)           # permet de créer une liste de 101 valeurs d'abscisses
y1_modele=[a1*i+b1 for i in x_modele]        # calcule l'ordonnée en appliquant le modèle  ai²+bi+c
y2_modele=[a2*i+b2 for i in x_modele]        # calcule l'ordonnée en appliquant le modèle  ai²+bi+c

# Tracer les graphique
plt.plot(x_modele,y1_modele,x_modele,y2_modele)
plt.plot(x_inter,y_inter,'go')
plt.plot(Vb,s,'r+')
plt.ylim(0,max(s))
plt.legend(('Modèle avant','modèle après','Equivalence','s(ms/cm)'))

plt.figure()
plt.plot(Vb1+Vb2,na,'g*')
plt.plot(Vb1+Vb2,nb,'r*')
plt.plot(Vb1+Vb2,nc,'b*')
plt.legend(('na','nb','nc'))
plt.show()
```
