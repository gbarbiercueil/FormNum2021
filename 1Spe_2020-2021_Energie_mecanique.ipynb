**<center>Calcul des √©nergies**  

On a d√©fini 3 listes, initialement vides :

- Ec pour l'√©nergie cin√©tique 
- Epp pour l'√©nergie potentielle de pesanteur
- Em pour l'√©nergie m√©canique

```python
%matplotlib notebook
from matplotlib import pyplot as plt

# Masse d'une balle de tennis en kilogramme et de g Newton par kilogramme
m=0.058
g=9.81

# D√©finition des listes
t=[0.04,0.08,0.12,0.16,0.2,0.24,0.28,0.32,0.36,0.4,0.44,0.48,0.52,0.56,0.6,0.64,0.68,0.72,0.76,0.8,0.84,0.88]     
x =[0.264,0.335,0.411,0.487,0.559,0.631,0.698,0.770,0.838,0.909,0.977,1.053,1.116,1.179,1.247,1.314,1.382,1.449,1.517,1.585,1.652,1.715]     
y=[0.538,0.702,0.846,0.972,1.078,1.17,1.242,1.301,1.339,1.364,1.377,1.373,1.347,1.314,1.263,1.196,1.116,1.017,0.901,0.774,0.631,0.471]
v=[4.727,4.267,3.870,3.436,3.058,2.693,2.387,2.121,1.912,1.804,1.797,1.780,1.746,1.946,2.242,2.500,2.783,3.174,3.491,3.771,4.131,4.435]
Ec=[]
Epp=[]
Em=[]

# D√©termination du nombre total de mesures
N=len(t)                 # len(t) donne longueur de la liste t (nombre de valeurs)

# On affiche la valeur minimale de la vitesse
print('la valeur minimale de la vitesse est : ',max(v),' m/s')      # max(v) donne la valeur minimale de la liste v

# Calcul des √©nergies        # Dans la boucle, les instructions indent√©es (d√©cal√©es vers la droite) sont r√©p√©t√©es
for i in range(0,N):         # pour i qui prend les valeurs 0, puis 1, puis... jusqu'√† N-1
    Ec.append(i)             # Ec.append(i) ajoute la valeur i √† la liste Ec
    Epp.append(-i)           # Epp.append(-i) ajoute la valeur -i √† la liste Epp
    Em.append(Ec[i]-Epp[i])  # Em.append(Ec[i]-Epp[i]) fait la diff√©rence entre les valeurs num√©rot√©es i des listes Ec et Epp

# Pr√©sentation des graphiques
plt.figure('Figure 1')
plt.plot(x,y,'ro')
plt.xlabel('x (en m)')
plt.ylabel('y en(m)')
plt.figure('Figure 2')
plt.plot(t,Ec,'r+',t,Epp,'b+',t,Em,'g*')
plt.xlabel('t (en s)')
plt.ylabel('E (en J)')
plt.legend(['?','?','?'])
plt.show()
```
